## Import the libraries

```python
import numpy as np
import pandas as pd
import scipy
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set() # this method overwrites all the styles and graphics of matplotlib graphs
import sklearn
```

## Load the Data

```python
data = pd.read_csv('../../Data/Simple-Linear-Regression.csv')
```

```python
data.describe() # describe is a Pandas method that displays the most useful descriptive statistics for each column in the data frame
```

Let's explore the problem: We have a sample of 84 students who studied in college.
Their total SAT = Critical Reading + Math + Writing, while their GPA = grade point average they have at graduation from university.
We will create a LR which predicts GPA based on the SAT score obtained
-- Set the SAT and get a score, with the score, you apply to college, the next 4 years, you attend the college and graduate, receiving many grades to form your final GPA

Each time you create a regression, It should be meaningful. Why would I predict GPA with SAT?  
    - SAT is considered one of the best estimators of intellectual capacity and capability
    - On average, if you did well in the SAT, you'll do well in college and in the work place
    - Almost all colleges across the USA are using SAT as a proxy for admission

## Create our first Regression

### Define the dependent and the Independent variables

```python
y = data['GPA']
x1 = data['SAT']
```

## Explore the data

```python
## Its always useful to plot your data in order to understand it better and to see if there is a relationship to be found
plt.scatter(x1, y)
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
```

Every point on the graph represents a student. We can see that there is strong relationship between SAT and GPA. In General, the higher the SAT of a student, the higher their GPA.

## Regression itself

```python
x = sm.add_constant(x1)
results = sm.OLS(y, x).fit()
results.summary()
```

```python
# Lets plot the regression line on the same scatter plot
plt.scatter(x1, y)
yhat = 0.2750 + 0.0017 * x1
fig = plt.plot(x1, yhat, lw=4, c='orange', label = 'Regression Line')
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
```

```python
# Lets plot the regression line on the same scatter plot
plt.scatter(x1, y)
yhat = 0.2750 + 0 * x1
fig = plt.plot(x1, yhat, lw=4, c='orange', label = 'Regression Line')
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
```
