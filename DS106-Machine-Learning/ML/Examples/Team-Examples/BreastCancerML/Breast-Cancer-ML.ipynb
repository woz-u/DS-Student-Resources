## task is to predict whether breast cancer is benign or malignant

```python
list(df.columns)
```

```python
df['diagnosis']
```

```python
df['diagnosis'].value_counts()
# 212 malignant case, 357 benign
```

```python
df.head()
```

```python
targets = np.where(df['diagnosis'] == 'M', 1,0)
```

```python
df['malignant'] = targets
```

```python
df.head()
```

```python
# drop diagnosis since malignant columns accounts
DF = df.drop(['diagnosis'], axis = 1)
```

```python
DF.head()
```

## Logistic Regression to predict malignancy

```python
data_with_targets = DF
```

```python
data_with_targets.shape
# data with targets has 32 columns & 569 rows
```

```python
# select inputs for regression
data_with_targets.iloc[:, :-1]
```

```python
unscaled_inputs = data_with_targets.iloc[:, :-1]
```

```python
unscaled_inputs.columns.values
```

## Standardize the data

```python
malignancy_scaler = StandardScaler()
```

```python
malignancy_scaler.fit(unscaled_inputs)
```

```python
scaled_inputs = malignancy_scaler.transform(unscaled_inputs)
```

```python
scaled_inputs
```

## Split data into train & test, then shuffle

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_inputs, targets, train_size = .7, random_state = 20)
# random state to shuffle observations in same way
```

```python
print(x_train.shape, y_train.shape)
```

```python
print(x_test.shape, y_test.shape)
```

## Logistic Regression with SkLearn

```python
# train the model
reg = LogisticRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
# evaluate model accuracy
reg.score(x_train, y_train)
```

```python
# In conclusion: utilizing logistic regression to predict breast cancer malignancy has an accuracy of just above 98%
# i.e.: the model learned to classify ~98% of the observations correctly
```

## Manually inspect accuracy of the model

## Testing the model

```python
# based on data the model has never seen before
reg.score(x_test, y_test)
# in 98% of the cases, the model predicted whether a diagnosis was malignant or not
```
