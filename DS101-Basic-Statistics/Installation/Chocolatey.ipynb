# Step 10: Install R kernel into Jupyter
a.  open a Command Prompt as administrator

b.  change directories by copying and pasting the below command into your Command Prompt

```shell
cd C:\Program Files\R\R-4.3.*\bin 
```

c. Then type R in the Command Prompt:

```shell
R
```

d. Rterm should open inside the command prompt/terminal:

e. Inside the Rterm, install the below commands in order:

```R
install.packages("devtools")
# select the closest CRAN mirror to your location

devtools::install_github("IRkernel/IRkernel")
# select 1 for All updates

IRkernel::installspec()

q()

y
```

Note: that if Rtools didn't install, go ahead and run:

```bash
install.packages("devtools", force=TRUE)
```
