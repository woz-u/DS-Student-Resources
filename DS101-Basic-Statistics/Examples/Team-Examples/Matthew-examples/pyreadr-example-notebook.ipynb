### Imports

```python
import pandas as pd
import pyreadr
```

### Reading In The RDA File

```python
result = pyreadr.read_r("Data/R-Datasets/Car.rda") # also works for Rds, rda

# result is a dictionary where keys are the name of objects and the values python
```

### Get Name of Dictionary

```python
print(result.keys())
```

### Creating a pandas DataFrame from the dictionary

```python
df1 = result["Car"]
df1.head()
```

### You can see how many rows your data has

```python
len(df1)
```

### You can check the shape of your data

```python
df1.shape
```

### Unless your screen is quite large, your output probably won’t display all 70 columns, let's change that!

```python
pd.set_option("display.max.columns", None)
df1.head()
```

### Showing Basics Statistics
Now that you’ve seen what data types are in your dataset, it’s time to get an overview of the values each column contains. You can do this with .describe():

```python
df1.describe()
```

### While it’s practical to see all the columns, you probably won’t need six decimal places, let's change it to two!

```python
pd.set_option("display.precision", 2)
df1.tail()
```
