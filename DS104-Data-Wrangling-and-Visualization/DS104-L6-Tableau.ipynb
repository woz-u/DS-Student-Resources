# Connecting to Multiple Datasets

The next step is learning how to pull in multiple datasets! There are two ways to do this:  

* **Joining:** Makes a new dataset by combining two separate ones using a key value pair.
* **Blending:** Leaves the datasets separate, but allows you to combine columns from multiple datasets in visuals with a temporary key.

You will only learn joining here; blending data is very rare. 

---

# Basic Concepts in Tableau

Before you go any further, you need to learn how Tableau breaks up the data.  

---

## Measures vs. Dimensions

First, you will see everything is broken down into *Dimensions* and *Measures*.  These are new names for old concepts.  Dimensions will be your Categorical Variables, and are thought of as ways to break up your data.  Measures are things that can be measured, and that will make them your Continuous Variables.  Tableau will automatically separate these, but will occasionally get one wrong.  You can manually change them by right clicking on them and selecting ```Convert to Dimension```, or ```Covert to Measure```.  

---

## Data Type

You will notice a little icon next to each variable, which denotes the what kind of variable Tableau thinks it is, whether that is ```#``` for number, ```Abc``` for string, ```T|F``` for Boolean, or something else entirely.  You will see that even though ```ID``` is considered a number with ```#```, Tableau is smart enough to put it in as a Dimension, because it is not something you can add or subtract, but rather something that you can use to separate out your data.

---

## Data Aggregation

Finally, Tableau will aggregate all measures.  By default, it will sum up the entire column into one big block. You can change how it is aggregated, by clicking the arrow on the side of it, selecting ```Measure(Sum)```, and selecting another option from the drop down menu.  However, no matter how you aggregate it, it will still just be one block.

![A window labeled Tableau Public - Book 1 has a menu bar with ten menus labeled file, data, worksheet, dashboard, story, analysis, map, format, window, and help. You need to add some dimension to break up the data.  This will always be true.  Which variables you add, the number of variables you add, and the options you select will determine the final chart and what you get out of it, but you will always need to divide your measures in some way or it will just be a clump of data.  Like frozen solid leftovers, you won't be able to separate out all your spaghetti strands.

## Types of References

There are a few different types of reference lines: 

* **Line**: A single line running through the graph
* **Band**: Highlight a particular section as particularly good or bad
* **Distribution**: Gradients based on a percentile
* **Box Plot**: A Box Plot.
