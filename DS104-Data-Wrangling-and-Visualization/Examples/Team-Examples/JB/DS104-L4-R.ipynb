# Part 3
### Use the following data to create a stacked bar chart using either Python or R.

```r
head(P3)
```

```r
# Basic Stacked Bar Chart
```

```r
ggplot(data= P3) +
  geom_bar(mapping = aes(x = Car, fill= Location)) + 
  ggtitle("Cars by Location") +
  xlab("Car") +
  ylab("Location") 
```

```r
# Bar Chart Filled
# Add postion= "fill" to fill to the top
```

```r
ggplot(data= P3) +
  geom_bar(mapping = aes(x = Car, fill= Location), position= "fill") + 
  ggtitle("Cars by Location") +
  xlab("Car") +
  ylab("Location") 
```

```r
# Bar Chart Multiple Variables
# Add postion= "dodge" to set varibles side by side
```

```r
ggplot(data= P3) +
  geom_bar(mapping = aes(x = Car, fill= Location), position= "dodge") + 
  ggtitle("Cars by Location") +
  xlab("Car") +
  ylab("Location") 
```

# Part 4
### The following dataset contains data for estuarine crocodiles, recording their head length and body length. Visualize the data with a clearly labeled scatter plot.

```r
head(P4)
```

```r
ggplot(data= P4) +
  geom_point(mapping = aes(x = HeadLength, y= BodyLength)) + 
  ggtitle("Crocodiles Head to Body Ratio") +
  xlab("Head Length") +
  ylab("Body Length")
```

```r
ggplot(data = P4, aes(x = HeadLength, y = BodyLength)) +
  geom_point() +
  geom_smooth(method='lm', se = TRUE)
```

```r
line <- lm(BodyLength ~ HeadLength, P4)
summary(line)

#
```

```r
ggplot(P5, aes(as.Date(P5$Month, `Heart Attacks`)) + geom_line() +
  scale_x_date() + xlab("") + ylab("Daily Heart Attacks") + ggtitle("Daily Heartattacks")
```
